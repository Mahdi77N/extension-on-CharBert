```python
text = """export CUDA_VISIBLE_DEVICES=1,2,3
DATA_DIR=data
MODEL_DIR=/kaggle/working/CharBERT/data/model/bert_base_cased
OUTPUT_DIR=model/output/mlm
python3 run_lm_finetuning.py \\
    --model_type bert \\
    --model_name_or_path ${{MODEL_DIR}} \\
    --do_train \\
    --do_eval \\
    --train_data_file $DATA_DIR/testdata/mlm_pretrain_enwiki.train.t \\
    --eval_data_file $DATA_DIR/testdata/mlm_pretrain_enwiki.test.t \\
    --term_vocab ${{DATA_DIR}}/dict/term_vocab \\
    --learning_rate {} \\
    --num_train_epochs 1 \\
    --max_steps {} \\
    --gradient_accumulation_steps {} \\
    --weight_decay {} \\
    --adam_epsilon {} \\
    --warmup_steps {} \\
    --max_grad_norm {} \\
    --char_vocab ./data/dict/bert_char_vocab \\
    --mlm_probability 0.10 \\
    --input_nraws 1000 \\
    --per_gpu_train_batch_size 1 \\
    --per_gpu_eval_batch_size 1 \\
    --save_steps 10000 \\
    --block_size 384 \\
    --overwrite_output_dir \\
    --mlm \\
    --output_dir ${{OUTPUT_DIR}}
""".format(learning_rate_, max_steps_, gradient_accumulation_steps_, weight_decay_, adam_epsilon_, warmup_steps_, max_grad_norm_)

with open('/kaggle/working/CharBERT/mlm.sh', 'w') as file:
    file.write(text)
```
